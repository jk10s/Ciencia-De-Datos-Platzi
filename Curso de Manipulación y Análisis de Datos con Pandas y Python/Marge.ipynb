```python
import pandas as pd
import numpy as np
```

```python
#marge es la union de dataframe atravez de sus indices 
#es decir que si continen los mismoa valores se pueden unir
```

```python
df_l=pd.DataFrame(
{'X':['']})
```

```python
df_left = pd.DataFrame(
                      {'X': ['x0', 'x1', 'x2', 'x3'],
                        'W': ['w0', 'w1', 'w2', 'w3'], 
                        'Y': ['y0', 'y1', 'y2', 'y3'] ,
                        'Mix': ['y2', 'y3', 'a2', 'a3']},
                       index=[0,1,2,3])
df_left
```

```python
df_right = pd.DataFrame(
                       {'Z': ['z2', 'z3', 'z4', 'z5'],
                         'A': ['a2', 'a3', 'a4', 'a5'], 
                         'Y': ['y2', 'y3', 'y4', 'y5']},
                        index=[2,3,4,5])
df_right
```

```python
#une los dataframe entes caso los y2 y y3
j=pd.merge(df_left, df_right)
j
```

```python
j.shape
```

```python
#especificar el tipo de union how
pd.merge(df_left, df_right, how = 'inner', on = 'Y')
```

```python
#union con mix izquierda y "y" de la derecha
pd.merge(df_left, df_right, how = 'inner', left_on = 'Mix', right_on = 'Y')
```

```python
pd.merge(df_left, df_right, how = 'inner', left_on = 'Mix', right_on = 'A')
```

```python
#tipos dde marge how
#inner funciona trayendo los mismo paramtros
#left se respeta los valores de izquierza y se añade los de derecha
#right se respeta los de derecha y se añade los de izquierda
#outer se uni todo
```

```python
#inner
pd.merge(df_left, df_right, how = 'inner', on = 'Y')
```

```python
#left se mantinen todos los de izquierza y se añaden dercha. los excedentes se añade un NaN
pd.merge(df_left, df_right, how = 'left', on = 'Y')
```

```python
#lright se mantinen todos los de derecha y se añaden izquierza. los excedentes se añade un NaN
pd.merge(df_left, df_right, how = 'right', on = 'Y')
```

```python
#outer se unificaron por los dos unicos valores en comun y2 y y3
pd.merge(df_left, df_right, how = 'outer', on = 'Y')
```
