```python
#agrupa dos indices el cut and color y saca el primedio del precio
df.groupby(['cut','color'])['price'].mean()
```

```python
#agregrar varias funciondes de un sola ve example min max and  mean
df.groupby(['cut','color'])['price'].aggregate(['min', np.mean,max])
```

```python
#funcion que recibe el agrupamiento y lo divide en 1000
def mean_kilo(x):
    return np.mean(x)/1000
```

```python
df.groupby(['cut','color'])['price'].aggregate(['min', np.mean,max,mean_kilo]).head(10)
```

```python
#diccionario con los indices de las columnas en este caso son dos carat el min y max y el price (mean, mean/100)
dict_agg={'carat':[min,max], 'price':[np.mean, mean_kilo]}
dict_agg
```

```python
#indices en las filsa pero ctbn en la s columnas
df.groupby(['cut','color']).aggregate(dict_agg)
```

```python
#funcion de filtro que saca los promedio oprencima de la media de 4
def filtro(x):
    return mean_kilo(x['price'])>4
```

```python
df.groupby('cut').filter(filtro)
```
