```python
anio_mes=lambda x: x[:7]
```

```python
historic["fecha_mes"]=historic["InvoiceDate"].astype(str).map(anio_mes)
```

```python
historic["Ventas"]= historic["UnitPrice"] * historic["Quantity"]
```

```python
historic.iloc[0:]
```

```python
historic = historic[historic["Ventas"]>0]
```

```python
historic
```

```python
ventas_mensuales = historic.groupby(["fecha_mes"]).agg({"Ventas":"sum"})
```

```python
ventas_mensuales
```
